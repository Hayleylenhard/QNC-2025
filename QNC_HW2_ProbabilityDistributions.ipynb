The paper that I chose (10.1016/j.cell.2017.05.018) did single-cell RNA-seq on microglia from AD and healthy mice and identified DAMs (disease associated microglia) as a microglia subtype present in the AD mice. They compare gene expression between healthy (homeostatic) microglia and DAMs.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_cells, n_genes = 200, 20
n_homeostatic = n_cells // 2
n_dam = n_cells - n_homeo

# Mean expression profiles
mean_homeostatic = np.zeros(n_genes)
mean_dam = np.concatenate([np.ones(5)*2, np.zeros(n_genes-5)])  # DAM upregulates 5 genes
cov = np.identity(n_genes) * 0.5  # Gaussian noise

# Simulate gaussian-distributed gene expression data
homeostatic = np.random.multivariate_normal(mean_homeostatic, cov, n_homeo)
dam = np.random.multivariate_normal(mean_dam, cov, n_dam)

# Pick one gene to visualize (ex: Gene 0 that differs between the homeostatic microglia and DAMs)
gene_idx = 0
homeostatic_gene = homeo[:, gene_idx]
dam_gene = dam[:, gene_idx]

# Plot histograms for expression values of gene 0 in the two subtypes
plt.hist(homeostatic_gene, bins=20, alpha=0.6, label="Homeostatic", color="blue", density=True)
plt.hist(dam_gene, bins=20, alpha=0.6, label="DAM", color="red", density=True)
plt.xlabel(f"Expression (Gene {gene_idx})")
plt.ylabel("Density")
plt.title("Gaussian-distributed expression: Homeostatic vs DAM")
plt.legend()
plt.show()
```
